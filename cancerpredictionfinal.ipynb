```python
#clean the data
sns.heatmap(data.isnull())
#The above line is used to visualize the presence of missing values (NaNs) in a dataset using a heatmap.
```

```python
#drop a colum
data.drop(["Unnamed: 32", "id"], axis=1, inplace=True)
data.head()
```

```python
#converting data to 1s and 0s
data.diagnosis=[1 if value=="M" else 0 for value in data.diagnosis]
```

```python
data.head()
```

```python
data["diagnosis"]=data['diagnosis'].astype("category",copy=False)
data["diagnosis"].value_counts().plot(kind="bar")
data.info()
data.head()
```

```python
#predictors are the input values in the data.
#target are the variables in the data.
data.diagnosis

#divide into target var and predictors
y=data["diagnosis"]#target var
x=data.drop(["diagnosis"] , axis=1)
```

```python
x #printed values in x
```

```python
y
```

## Normalizing the Data

```python
#normalise data : the predictors into the model have diff units makes all the units normalized
#scikit for ML 
from sklearn.preprocessing import  StandardScaler 
#crete a scaler object
scaler= StandardScaler()
#() initializing an instance of the class
#fit the scalar to the data and tranform the data
x_scaled=scaler.fit_transform(x)
x_scaled
```

## Split  the Data

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.30, random_state=42)
#train_test_split takes predictors,target value , proportion , random state is the arbitary number , if you ever need to repeat the split  you will have to pass the same number to get the same division 
```

## Train The Model

```python
from sklearn.linear_model import LogisticRegression
#create Logistic regression model
lr=LogisticRegression()

#train the training data
lr.fit(x_train,y_train)

#predict the target variable based on the test data
y_prediction = lr.predict(x_test)
```

## Evaluation of the Model

```python
y_prediction
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
accuracy=accuracy_score(y_test,y_prediction)
print(f"Accuracy :{accuracy:.2f}")
#0.98 is qual to 98%
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_prediction))
# Value of 1 indicates Malignant, meaning the patient has cancer.
# Value of 0 indicates Non Malignant , Meaning patient dosent has cancer
```
